```python
import requests
import json
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
API_KEY = ""
with open("apikey.json") as f:
    content = f.read()
    API_KEY = json.loads(content)['apikey']
```

## Get recipe data from Spoonacular API

```python
# number of recipes

n_recipes = 10 # unit: hundred, e.g. 10 hundred = 1000
```

```python
def get_100_random_recipes():
    '''
    This function requests 100 random recipes from Spoonacular 
    100 is the highest number of recipes one could get from a single request
    
    return: List( Dict() ) - a list of recipes
    '''
    res = requests.get(f"https://api.spoonacular.com/recipes/random?apiKey={API_KEY}&number=100")
    recipes = json.loads(res.content)["recipes"]
    return recipes
```

```python
def get_random_recipes(n):
    '''
    This function called get_100_random_recipes 10 times 
    to get 1000 random recipes in total
    and create a DataFrame with the data
    
    return: pd.Dataframe
    '''
    recipes = []
    for _ in range(n):
        recipes_100 = get_100_random_recipes()
        recipes+=recipes_100
    df = pd.DataFrame(recipes)
    return df
```

```python
# get data from spoonacular
recipe_data = get_random_recipes(n_recipes)
```
